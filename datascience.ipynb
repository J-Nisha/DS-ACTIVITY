```python
#Write python code for indexing rows and columns based on the following conditions:
#Assume we have the following dataframe:
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],

       'age': [25, 35, 40, 28],

      'gender': ['F', 'M', 'M', 'M'],

      'salary': [50000, 70000, 60000, 80000]}

df = pd.DataFrame(data)
print(df)
```

```python
#a. Select rows where age is greater than 30
age_df = df[df['age'] > 30]
print(age_df)

```

```python
#b. Select rows where name contains 'e':
fd_df = df[df['name'].str.contains('e')]


print(fd_df)
```

```python
#c. Select rows where gender is 'M' and salary is greater than 65000:
salary_df = df[(df['gender'] == 'M') & (df['salary'] > 65000)]

print(salary_df)
```

```python
#d. Select columns 'name' and 'age'
print(df[['name', 'age']])
```

```python
#a.  select the rows where clients with primary education have subscribed to a deposit?
sub_df = df[(df['education'] == 'primary') & (df['deposit'] == 'yes')]
print(sub_df)
```

```python
#select the rows where clients who have not subscribed to a deposit?
de_df = df[(df['deposit'] == 'no')]
print(de_df)
```

```python
#c. select the rows where clients who have subscribed to a deposit either have a housing or a personal loan?
pe_df = df[(df['deposit'] == 'yes') & ((df['housing'] == 'yes') | (df['loan'] == 'yes'))]


print(pe_df)
```

```python
#d. select the rows where clients with secondary education who have not subscribed to a deposit?
filtered_df = df[(df['education'] == 'secondary') & (df['deposit'] == 'no')]

print(filtered_df)
```

```python
#e. select the rows where  clients who have subscribed to a term deposit as an outcome of the successful marketing campaign?
filtered_df = df[(df['deposit'] == 'yes') & (df['poutcome'] == 'success')]

print(filtered_df)
```

```python
#f. select the rows where unemployed clients who have not subscribed to deposit?
unemployed_no_deposit = df[(df['job'] == 'unemployed') & (df['deposit'] == 'no')]
print(unemployed_no_deposit)
```

```python
#g) select coloumns ‘education’ and ‘balance’ where age is less than or equal to 30
young_clients = df.loc[df['age'] <= 30, ['education', 'balance']]
print(young_clients)
```
